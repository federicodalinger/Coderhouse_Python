```python
"""
>> Consigna: Crear un programa para calcular la nota final del estudiante en base a dos exámenes,
los exámenes cuentan con un porcentaje distinto de la nota final
nota_1  cuenta como el 40% de la nota final
nota_2 cuenta como el 60% de la nota final
>> Aspectos a incluir en el entregable:
Tener en cuenta los temas vistos en la clase 1: números, print, input, variables, operaciones matemáticas, cadena de texto. 
>> Aspectos a tener en cuenta:
Los datos deben guardarse en variables y deben ser dinámicos por medio de input.
"""
#Creo variable con nombre de alumno:
alumno=str(input("Ingrese el nombre del alumno: "))

#Defino los % que cuentan de cada nota para la nota final:
porcent_nota_1=40
porcent_nota_2=60

#Creo variable para cada nota a ingresar (indico flot porque puede ser decimal, ej: 9.5):
nota_1=float(input(f"Ingrese la 1er nota de {alumno}: "))
nota_2=float(input(f"Ingrese la 2da nota de {alumno}: "))

#Creo variable de la nota final (aplicando operacion suma, multiplicacion y division):
nota_final=(porcent_nota_1*nota_1+porcent_nota_2*nota_2)/100

#Muestro la nota final del alumno en cuestion, segun sus notas 1 y 2 ingresadas.
#Uso ":.2f" para limitar la cantidad de decimales a mostrar, sin redondear:
print(f"La nota final de {alumno} es: {nota_final:.2f}")
```
