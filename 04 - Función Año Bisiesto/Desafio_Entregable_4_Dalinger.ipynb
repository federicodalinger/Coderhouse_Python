```python
# Definimos la funcion "bisiesto":
def bisiesto(numero_ingresado):

  # Si el valor ingresado es un numero, entero, entra en el condicional:
  if ( type(numero_ingresado) == int):

    # Se plantean las condiciones para ser y no año bisiesto:
    if (int(numero_ingresado)%400 == 0):
      mensaje = print (f"El año {numero_ingresado} es bisiesto (múltiplo de 400).")
      return mensaje
    elif (int(numero_ingresado)%100 == 0):
      mensaje = print (f"El año {numero_ingresado} NO es bisiesto (múltiplo de 100).")
      return mensaje
    elif (int(numero_ingresado)%4 == 0):
      mensaje = print (f"El año {numero_ingresado} es bisiesto (múltiplo de 4).")
      return mensaje
    else:
      mensaje = print(f"El año {numero_ingresado} NO es bisiesto (no es múltiplo ni de 4 ni de 400).")
      return mensaje 
  
  # Si el valor ingresado NO fuese un numero, indica el error de ingreso:
  else: 
    mensaje = print(f"'{numero_ingresado}' no es un número, favor de ingresarlo de nuevo.") 
    return mensaje
```

```python
#Probamos la funcion con los ejemplos del ejercicio:

bisiesto(2012)
bisiesto(2010)
bisiesto(2000)
bisiesto(1900)
bisiesto("text")
bisiesto("")
bisiesto("    ")
```
